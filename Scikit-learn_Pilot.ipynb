## Preprocessing

```python
# StandardScaler is like standard normal distribution (also called Z-score normalization)
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

scaler = StandardScaler() 
# creates a scaler object (like getting a measuring tool)
# scaler is an instance of the class StandardScaler
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
# fit: learn the mean and std from training data
# transform: apply the scaling to training data using learned values
# only transform the test data using learned values
# never fit / learn from test data ‚Äì prevents data leakage!
X_train_scaled # look at the scaled data, all values are roughly on the same scale
```

```python
'''formula: (x ‚Äì mean) / std, to do it manually

import numpy as np
((X_train ‚Äì np.mean(X_train, axis = 0)) / np.std(X_train, axis = 0))
# axis = 0 means column-wise operation

# MinMaxScaler scales data to a fixed range (usually 0 to 1)
# formula: (x ‚Äì min) / (max ‚Äì min) 

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# rule of thumb: measures distances or uses gradients (scale); rule-based split (no scale)'''
```

## Classification

```python
from sklearn.datasets import load_breast_cancer 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

X, y = load_breast_cancer(return_X_y=True) 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 
scaler = StandardScaler() 
X_train_scaled = scaler.fit_transform(X_train) 
X_test_scaled = scaler.transform(X_test)  
clf = KNeighborsClassifier()
clf.fit(X_train_scaled, y_train)
clf.score(X_test_scaled, y_test)
```

## Regression

```python
# classification predicts categories (discrete labels): cat vs dog, spam vs not spam
# whereas regression predicts continuous values: house prices, temperature, age

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

X, y = fetch_california_housing(return_X_y=True) 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler = StandardScaler() 
X_train_scaled = scaler.fit_transform(X_train) 
X_test_scaled = scaler.transform(X_test) 

reg = LinearRegression()
reg.fit(X_train_scaled, y_train)

# this is not accuracy, this is R2-score: measures how well the regression model explains the variance in the target variable
reg.score(X_test_scaled, y_test)
```

```python
sample = X_test_scaled[0]
reg.predict([sample]) # predicting median house value # 1.17 mean $75k
```

## Clustering

#### definition: automatically discovers hidden patterns and groups similar data points together without any pre-existing labels.

```python
from sklearn.datasets import make_blobs # generation func
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X, _ = make_blobs(n_samples=5000, centers=5, random_state=10)
# we use _ when we don‚Äôt need the variable
# n_samples = data points, centers = num. of natural clusters
# random_state = produces same / consistent results

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

import matplotlib.pyplot as plt
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
# visualization before clustering
# [:, 0] ‚Äì all rows, first feature
# [:, 1] ‚Äì all rows, second feature
```

```python
kmeans = KMeans(n_clusters=5) # init KMeans algo looking for 5 clusters
kmeans.fit(X_scaled) # run clustering on scaled data
# algo finds cluster centers and assign each point to nearest center

plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_)
# c=kmeans.labels_: color each point according to its cluster assignment (0-4)
```

```python
from sklearn.datasets import make_moons
X, _ = make_moons(n_samples=5000, noise=0.08) 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
```

## PCA

#### Principal Component Analysis: a dimensionality reduction technique that converts original features into new, uncorrelated features that capture the most important pattern in data.

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X, y = fetch_openml('mnist_784', return_X_y=True) 
# MNIST handwritten digits dataset
# x: 70,000 images of 28x28 pixels (flattened to 784 features)
# y: labels (0-9) for each image

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# 80% training (56,000); 20% testing (14,000) # random_state=0 produces same result

pca = PCA(n_components=20) # reduces features from 784 to 20
X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
X_train.shape
```

```python
X_train_reduced.shape
```

```python
# with 784 features
clf = LogisticRegression(max_iter=100)
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
# with 20 features
clf = LogisticRegression(max_iter=100)
clf.fit(X_train_reduced, y_train)
print(clf.score(X_test_reduced, y_test))
```

```python
# accuracy is still good enough with just 1.3% of the original features, ‚Äòcause PCA captures the most important visual patterns
```

## Hyperparameter Tuning

#### finding the best settings for ML algo (classification, regression, clustering, dimensionality reduction, reinforcement learning, etc.) to maximize performance!

```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
param_grid = {			          
    'n_estimators': [50, 100, 200],	   # num. of trees in forest	
    'min_samples_split': [2, 5],		# min samples to split a node
    'max_depth': [None, 5, 10]		# max depth of trees (None = unlimited)
} # all possible combinations: 3 x 2 x 3 = 18 diff parameter sets

clf = RandomForestClassifier(n_jobs=-1) # n_jobs=-1 uses all CPU cores
grid = GridSearchCV(clf, param_grid, cv=3) # 18 x 3 = 54 total model trainings 
grid.fit(X_train, y_train)
```

```python
grid.best_params_ # these param set gives the highest average cross-validation score
```

```python
best_clf = grid.best_estimator_ # RandomForest with the optimal parameters
```

```python
best_clf.score(X_test, y_test) # evaluate on unseen test data
```

## Pipelines

#### a way to chain multiple processing steps together in a single object, where the output of each step becomes the input to the next step!

```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

pipe = Pipeline([
    ('scale', StandardScaler()), # scaling the data
    ('pca', PCA(n_components=10)), # reducing dimensions
    ('forest', RandomForestClassifier()) # training model
])

pipe.fit(X_train, y_train)
pipe.score(X_test, y_test) # accuracy
```

```python
y_pred = pipe.predict(X_test) 
y_pred # prediction
```
